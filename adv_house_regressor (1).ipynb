```python
import pandas as pd
```

```python
data = pd.read_csv("house-price/train.csv")
data.head()
```

```python
x = data.drop("SalePrice" , axis = 1)
y = data["SalePrice"]
```

```python
def object_type(x):
    l = []
    for i in x.columns:
        if x[i].dtype == "object":
            l.append(i)
    return l
obj_col = object_type(x)
obj_col
```

```python
def empty_object(obj_col):
    empty = []
    for i in obj_col:
        if data[i].isna().sum() > 0 :
            empty.append(i)
    return empty
e_col = empty_object(obj_col)
e_col
```

```python
def dict_ob(e_col):
    dic = {}
    for i in e_col:
        dic[i] = data[i].isna().sum()
    return dic
dict_ob = dict_ob(e_col)
dict_ob
```

```python
from sklearn.impute import SimpleImputer
absence_cols = e_col
for col in absence_cols:
    if col in x.columns:
        x[col] = x[col].fillna("None")

obj_cols = obj_col

true_cat_cols = [col for col in obj_cols if col not in absence_cols]


for col in true_cat_cols:
    missing_ratio = x[col].isna().mean()

    if missing_ratio < 0.3:
        imputer = SimpleImputer(strategy="most_frequent")
        x[col] = imputer.fit_transform(x[[col]]).ravel()
    else:
        x[col] = x[col].fillna("Unknown")
```

```python
import numpy as np

numeric_cols = x.select_dtypes(include=[np.number]).columns.tolist()

num_imputer = SimpleImputer(strategy="median")
x[numeric_cols] = num_imputer.fit_transform(x[numeric_cols])
x.isna().sum().sort_values(ascending=False).head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cat_cols = X_train.select_dtypes(include="category").columns
num_cols = X_train.select_dtypes(exclude="category").columns

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)
X_train_transformed = preprocessor.fit_transform(X_train)
X_val_transformed = preprocessor.transform(X_val)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

## RANDOM FOREST — GridSearchCV

```python
rf = RandomForestRegressor(
    random_state=42,
    n_jobs=-1
)

rf_param_grid = {
    "n_estimators": [300, 500],
    "max_depth": [None, 20, 30],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt"]
}

rf_grid = GridSearchCV(
    estimator=rf,
    param_grid=rf_param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5,
    n_jobs=-1,
    verbose=1
)

rf_grid.fit(X_train_transformed, y_train)
```

## EXTRA TREES — GridSearchCV

```python
et = ExtraTreesRegressor(
    random_state=42,
    n_jobs=-1
)

et_param_grid = {
    "n_estimators": [500, 800],
    "max_depth": [None, 20, 30],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt"]
}

et_grid = GridSearchCV(
    estimator=et,
    param_grid=et_param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5,
    n_jobs=-1,
    verbose=1
)

et_grid.fit(X_train_transformed, y_train)
```

## HIST GRADIENT BOOSTING — GridSearchCV

```python
X_train_dense = X_train_transformed.toarray()
X_val_dense   = X_val_transformed.toarray()
hgb = HistGradientBoostingRegressor(
    random_state=42
)

hgb_param_grid = {
    "learning_rate": [0.03, 0.05],
    "max_depth": [5, 8],
    "max_leaf_nodes": [31, 63],
    "min_samples_leaf": [20, 30],
    "l2_regularization": [0.0, 0.1]
}

hgb_grid = GridSearchCV(
    estimator=hgb,
    param_grid=hgb_param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5,
    n_jobs=-1,
    verbose=1
)
hgb_grid.fit(X_train_dense, y_train)
```

```python
print("Best CV RMSE:")
print("RandomForest:", -rf_grid.best_score_)
print("ExtraTrees  :", -et_grid.best_score_)
print("HistGB      :", -hgb_grid.best_score_)
```

```python
y_train_log = np.log1p(y_train)
y_val_log   = np.log1p(y_val)
hgb_grid.fit(X_train_dense, y_train_log)
preds_log = hgb_grid.best_estimator_.predict(X_val_dense)
preds = np.expm1(preds_log)

rmse = np.sqrt(mean_squared_error(y_val, preds))
print("Validation RMSE (log target):", rmse)
```

## log is rejected to to high rmse

```python
test_data = pd.read_csv("house-price/test.csv")
X_test = test_data.copy()

X_test_transformed = preprocessor.transform(X_test)
X_test_dense = X_test_transformed.toarray()
test_preds_log = hgb_grid.best_estimator_.predict(X_test_dense)
test_preds = np.expm1(test_preds_log)
```
